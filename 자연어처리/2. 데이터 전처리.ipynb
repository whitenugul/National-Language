# 2. 토큰화(Tokenization)
* 전처리 과정에는 토큰화, 정제, 정규화 작업등이 있다.
* 이때 토큰화는 원시 데이터를 가져와서 유용한 데이터 문자열로 변환하는 간단한 프로세스를 말한다.
* 사이버 보안, NFT 생성에 사용되는 것으로 유명, 자연어 프로세스의 중요한 부분을 차지하게 되었다.
* 토큰의 단위는 자연어 내에서 의미를 가지는 최소 단위로 정의
* 단락과 문장을 보다 쉽게 의미를 할당할 수 있는 더 작은 단위로 분할하는데 사용한다.

> 토큰화 작업은 주어진 코퍼스 내 자연어 문장들을 토큰이라 불리는 최소 단위로 나누는 작업이라고 할 수 있다.

### 2-1. 토큰화 과정의 필요성
* 언어 모델의 자연어 이해 능력 향상
* 다양한 자연어를 효율적으로 표현이 가능하다.
  - 중복을 제거한 대규모 자연어 코퍼스 내 토큰의 집합(단어 사전) 사용
  - 적절한 정보량을 내포하면서 전체 단어 사전의 개수가 많아지지 않도록 토큰의 단위를 잘 정의해야 한다.
  - 상황에 따라 다르지만, 일반적인 자연어처리 작업에서 단어 사전의 규모는 약 10,000개 ~ 50,000개 정도로 구성(현대 한글의 글자 수는 11,172개)


### 2-2. 토큰화 방법
* 문장 토큰화: 토큰의 기준을 문장으로 하는 토큰화 방법
  - 문장의 끝에 오는 문장 부호를 기준으로 코퍼스를 잘라냄(. 또는 ! 또는 ?)
  - 예외) 제 이메일 주소는 andrewahn00@gmail.com입니다.에서 gmail과 com 사이에 마침표에서 잘릴 수도 있다.
* 단어 토큰화: 토큰의 기준을 단어로 하는 토큰화 방법
  - 보편적으로 구분기호를 가지고 텍스트를 나누게 되며, 기본적으로 공백을 구분자로 사용
  - 한국어의 경우 교착어이기 때문에 공백으로 단어 토큰화를 하면 성능이 좋지는 않다.
  - 새로운 단어가 추가될수록 단어 사전의 크기가 증가한다.
  - OOV(Out Of Vocabulary) 문제
* 문자 토큰화: 토큰의 기준을 문자로 하는 토큰화 방법
  - 단어 토큰화의 한계점들을 해결하기 위한 방법
  - 영어는 26개의 알파벳에 따라 분리, 한국어는 자음 19개와 모음 21개의 글자에 따라 분리된다.
  - 문장 하나를 생성하는데 너무 많은 추론이 필요하다
  - 단어 사전은 작지만 모델의 예측 시간에 문제가 생길 수 있다.
* 서브워드 토큰화: 토큰의 기준을 서브워드로 하는 토큰화 방법
  * 서브워드는 알고리즘이 있다. => 구글에서 제작함
  - 단어 토큰화와 문자 토큰화의 한계점을 해결하기 위한 방법
  - 문자 토큰화의 확장된 버전으로 토큰의 단위를 n개의 문자로 정의하고 해당 기준에 따라 텍스트를 분리하는 방법
  - 형태소 분절 기반의 서브 워드 토큰화로 확장될 수 있어 한국에서도 좋은 성능을 가진다.
  -서브 워드를 만드는 알고리즘 중에서 가장 유명한 것이 BPE


### 3-1. 서브워드(subword)
* 단어보다 더 작은 의미의 단위
* 예) Birthday = birth + day, 아침밥 = 아침 + 밥
* 단어를 여러 서브워드로 분리해서 단어 서전을 구축하겠다는 토큰화 방법이다.
* 신조어에서 주로 발생하는 OOV 문제를 완화
* upperwear

### 3-2. BPE(Byte Pair Encoding)
* 코퍼스 내 단어의 등장 빈도에 따라 서브워드를 구축하는데 사용한다.
* BPE 알고리즘은 2016년 [Neural Machine Translation of Rare Words with Subword Units](https://arxiv.org/abs/1508.07909) 논문에서 처음 제안 되었다.
* 글자 단위에서 점진적으로 서브 워드 집합을 만들어내는 Bottom-up 방식의 접근 방식으로 자연어 코퍼스 내에 있는 모든 단어들을 글자 단위로 분리한 뒤에, 등장 빈도에 따라 글자들을 서브워드로 통합하는 방식이다.


### 3-3. WordPiece Tokenizer
* 구글이 2016년도 [Google's Neural Machine Translation System Bridging the Gap between Human and Machine Translation](https://arxiv.org/abs/1609.08144) 논문에서 처음 공개한 BPE의 변형 알고리즘이다.
* 병합할 때 두 문자가 있을 때, 각각의 문자가 따로 있을 때를 더 중요시 여기는지, 병합되었을 때를 더 중요시 여기는지에 차이점을 둔다.
* GPT 모델과 같은 생성 모델의 경우에는 BPE 알고리즘을 사용한다.
* BERT, ELECTRA와 같은 자연어 이해 모델에서는 WordPiece Tokenizer를 주로 사용한다.

# 5. 정규화(Normalization)
* 일반적인 머신러닝 작업에서 데이터 정규화는 학습 데이터의 값들이 적당한 범위를 유지하도록 데이터의 범위를 반환하거나 스케일링하는 과정
* 정규화 목표는 모든 데이터가 같은 정도의 스케일로 반영되도록 하는 것이다.
* 자연어 처리 정규화의 핵심은 표현 방법이 다른 단어들을 통합시켜서 같은 단어로 만들어 주는 과정이다.

### 5-1. 정규화 작업이 필요한 이유
* 이상적으로 단어 사전내의 단어 토큰들이 모두 중요하게 고려되길 원한다.
* 자연어의 특성상 의미가 같은데 표기가 다른 단어들이 있을 수 있고, 의미는 같지만 사용 빈도가 낮은 동의어들은 학습에 유용하게 활용되지 않을 수 있다.
* 의미가 같지만 표기가 다른 단어들을 통합한다면 통합된 단어의 사용빈도가 높아질 것이고, 빈도가 낮은 단어들의 중요도가 높아질 수 있다.



### 5-2. 정규화 작업의 종류
* 어간 추출, 표제어 추출
  - 어간 추출: 형태학적 분석을 단순화한 버전으로, 정해진 규칙만 보고 단어의 의미를 자르는 어림짐작의 작업
  - 표제어 추출: 단어들이 다른 형태를 가지더라도, 그 뿌리 단어를 찾아가서 단어의 개수를 줄일 수 있는지 판단하는 방법을 말한다.
    - 예시) am, are, is의 뿌리 단어는 be이다.
* 대소문자 통합
  - 대문자와 소문자가 구분되어야 하는 경우도 있다.
  - 그러므로 무작정 소문자로 통합해서도 안된다.
  - 예) US(미국), us(우리) / Bush(미국 대통령 이름), bush(수풀)

### 5-3. 정규화 시 유의할 점
* 규칙이 너무 엄격한 정규화 방법은 부작용이 심해 학습에 악영향을 줄 수 있다.
* 원본의 의미를 최대한 유지하는 것이 학습에 도움이 된다.
* 대화에서 사용하는 의미가 비슷한 이모티콘들을 통합하는 정규화 작업
  - 'ㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋ' => 'ㅋㅋ'

### 7-1. 불용어 제거

```python
# 불용어 사전 정의
stopwords = ['이하', '바로', '☞', '※', '…']

def delete_stopwords(context):
  preprocessed_text = []
  for text in context:
    text = [w for w in text.split(' ') if w not in stopwords] # 불용어가 들어있지 않은 문장들만 list에 담아 text라는 변수로 보낸다.
    preprocessed_text.append(' '.join(text))
  return preprocessed_text
```

```python
processed_context = delete_stopwords(context)
```

```python
for _, text in enumerate(processed_context):
  print(text)
```

### 7-3. 문장 분리
* 학습 데이터를 구성할 때 입력 데이터의 단위를 설정하기 애매해지므로 문장 단위로 모델이 학습하도록 유도하기 위해 문장 분리가 필요하다
* 한국어 문장 분리기 중, 가장 성능이 우수한 것으로 알려진 kss 라이브러리를 사용한다.
* https://github.com/hyunwoongko/kss

```python
!pip install kss
```

```python
import kss
```

```python
def sentence_separator(processed_context):
  splited_context = []
  for text in processed_context:
    text = text.strip()
    if text:
      splited_text = kss.split_sentences(text)
      # split_sentences에 의해 list로 분리된다.
      splited_context.extend(splited_text)
      # splited_text 안의 리스트의 모든 내용이 splited_context로 들어간다.
  return splited_context
```

```python
splited_context = sentence_separator(processed_context)
```

```python
for i, text in enumerate(splited_context):
  print(i, text)
```

### 7-4. 이메일 제거

```python
def delete_email(context):
    preprocessed_text = []
    for text in context:
        text = re.sub(r'[a-zA-Z0-9+-_.]+@[a-zA-Z0-9-]+\.[a-zA-Z0-9-.]+', '', text).strip()
        if text:
            preprocessed_text.append(text)
    return preprocessed_text
```

```python
preprcessed_context = delete_email(processed_context)
```

### 7-6. 저작권 관련 택스트 제거

```python
def delete_copyright(context):
    re_patterns = [
        r'\<저작권자(\(c\)|©|(C)|(\(C\))).+?\>',
        r'(Copyrights)|(\(c\))|(\(C\))|©|(C)|'
    ]
    preprocessed_text = []
    for text in context:
        for re_pattern in re_patterns:
            text = re.sub(re_pattern, "", text).strip()
        if text:
            preprocessed_text.append(text)
    return preprocessed_text
```

```python
mecab = Mecab()
morphs = mecab.pos('아버지가방에들어가신다', join=False)
print(morphs)
```

```python
# 형태소 출력
for text in normalized_context:
  morph = mecab.pos(text, join=False)
  for m in morph:
    pos_tag = m[1]
    print(pos_tag)
```

```python
# 명사(NN), 동사(V), 형용사(J)의 포함 여부에 따라 문장 필터링
def morph_filter(context):
    NN_TAGS = ['NNG', 'NNP', 'NNB', 'NP']
    V_TAGS = ['VV', 'VA', 'VX', 'VCP', 'VCN', 'XSN', 'XSA', 'XSV']
    J_TAGS = ['JKS', 'J', 'JO', 'JK', 'JKC', 'JKG', 'JKB', 'JKV', 'JKQ',
                'JX', 'JC', 'JKI', 'JKO', 'JKM', 'ETM']
    preprocessed_text = []
    for text in context:
        morphs = mecab.pos(text, join=False)
        nn_flag = False
        v_flag = False
        j_flag = False
        print(morphs)
        for morph in morphs:
            pos_tags = morph[1].split("+") # 1개만 있는 경우도 있지만 +로 묶여있는 경우도 있다. => +가 없더라도 split을 해도 무방하다.
            for pos_tag in pos_tags:
                if not nn_flag and pos_tag in NN_TAGS: # 명사(NN)가 있는 경우 True로 바뀐다.
                    nn_flag = True
                if not v_flag and pos_tag in V_TAGS: # 동사(V)가 있는 경우 True로 바뀐다.
                    v_flag = True
                if not j_flag and pos_tag in J_TAGS: # 형용사(J)가 있는 경우 True로 바뀐다.
                    j_flag = True
                # not nn_flag는 nn_flag가 False일때 True가 된다.
                # not v_flag, not j_flag도 마찬가지
            if nn_flag and v_flag and j_flag:# 명사, 동사, 형용사가 전부 존재하는 경우(nn_flag, v_flag, j_flag가 전부 True인 경우) preprocessed_text안에 text가 들어가게 된다.
                preprocessed_text.append(text)
                break
    return preprocessed_text
```

```python
post_processed_context = morph_filter(normalized_context)
```

```python
for i, text in enumerate(post_processed_context):
  print(i, text)
```

### 7-10. 문장 길이 기반 필터링

```python
def min_max_filter(min_len, max_len, context):
  post_processed_context = []
  for text in context:
    if len(text) > min_len and len(text) < max_len:
      post_processed_context.append(text)
  return post_processed_context
```
