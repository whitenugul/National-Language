```python
AI_text = [word.replace('\t', '') for word in AI_text]
```

```python
new_df = pd.DataFrame({'document':AI_text})
```

```python
new_df
```

```python
new_df.replace('', float('NaN'), inplace=True)
print(new_df.isnull().values.any())
```

```python
new_df = new_df.dropna().reset_index(drop=True)
print(f'필터링된 데이터셋 총 개수: {len(new_df)}')
```

```python
def erase_special(text):
    processed_text = re.sub(r"[^\w\s]", " ", text)
    return processed_text

new_df['document'] = new_df['document'].map(erase_special)
```

```python
new_df
```

```python
def left_ko(text):
    hangul = re.compile('[^ ㄱ - ㅣ 가-힣+]')
    result = hangul.sub('', text)
    return(result)

new_df['document'] = new_df['document'].map(left_ko)
```

```python
new_df
```

```python
new_df['document'] = new_df['document'].apply(lambda x: ' '.join([token for token in x.split() if len(token) > 2]))
new_df
```

```python
new_df = new_df[new_df.document.apply(lambda x: len(str(x)) > 200 and len(str(x).split()) > 5)].reset_index(drop=True)
new_df
```

```python
mecab = Mecab()

def tokenizer_stopword_dic(text):
    nouns = mecab.nouns(text)
    return [noun for noun in nouns if noun not in stopword]

new_df['document'] = new_df['document'].map(tokenizer_stopword_dic)
```

```python
new_df
```

```python
# tokenized_doc = new_df['document'].apply(lambda x: x.split())
tokenized_doc = new_df['document'].apply(lambda x: [s_word.strip() for s_word in x if s_word.strip() not in stopword])
```

```python
tokenized_doc
```

```python
tokenized_doc = tokenized_doc.to_list()
print(len(tokenized_doc))
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(tokenized_doc)
```

```python
word2idx = tokenizer.word_index
```

```python
idx2word = {value: key for key, value in word2idx.items()}
```

```python
encoded = tokenizer.texts_to_sequences(tokenized_doc)
```

```python
vocab_size = len(word2idx) + 1
print(f'단어 사전의 크기: {vocab_size}')
```

```python
# negative sampling을 위해 keras에서 제공하는 전처리 도구인 skipgrams를 사용한다.
from tensorflow.keras.preprocessing.sequence import skipgrams
```

```python
skip_grams = [skipgrams(sample, vocabulary_size=vocab_size, window_size=10) for sample in encoded[:5]]
# vocabulary_size는 negative_sampling을 진행할 개수를 넣을 수 있다.

print(f'전체 샘플 수 : {len(skip_grams)}')
```

```python
pairs, labels = skip_grams[0][0], skip_grams[0][1]
print(f'3pairs : {pairs[:3]}')
print(f'3labels : {labels[:3]}')
```

```python
for i in range(5):
  print('({:s} ({:d}) {:s} ({:d}) -> {:d})'.format(
      idx2word[pairs[i][0]], pairs[i][0],
      idx2word[pairs[i][1]], pairs[i][1],
      labels[i]
  ))
```

```python
training_dataset = [skipgrams(sample, vocabulary_size=vocab_size, window_size=10) for sample in encoded[:10000]]

print(training_dataset[0])
```

```python
print(len(training_dataset))
```

```python
# word embedding 구축
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Embedding, Reshape, Activation, Input, Dot
from tensorflow.keras.utils import plot_model
```

```python
embedding_dim = 100 # weight를 100으로 준다.

# 중심 단어를 위한 Embedding table
w_inputs = Input(shape=(1,), dtype='int32')
word_embedding = Embedding(vocab_size, embedding_dim)(w_inputs)

# 주변 단어를 위한 Embedding table
c_inputs = Input(shape=(1,), dtype='int32')
context_embedding = Embedding(vocab_size, embedding_dim)(c_inputs)
```

```python
dot_product = Dot(axes=2)([word_embedding, context_embedding])
dot_product = Reshape((1,), input_shape=(1, 1))(dot_product)
output = Activation('sigmoid')(dot_product)

model = Model(inputs=[w_inputs, c_inputs], outputs=output)
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
for epoch in range(100):
    loss = 0
    for _, elem in enumerate(training_dataset):
        first_elem = np.array(list(zip(*elem[0]))[0], dtype='int32')
        second_elem = np.array(list(zip(*elem[0]))[1], dtype='int32')
        labels = np.array(elem[1], dtype='int32')
        X = [first_elem, second_elem]
        Y = labels
        loss += model.train_on_batch(X, Y)
    print('Epoch:', epoch+1, 'Loss: ', loss)
```
